```python
import pandas as pd
import re

# Charger le fichier CSV
file_path = r"C:\Users\Utilisateur\Downloads\infos_especes (1).csv"
data = pd.read_csv(file_path, delimiter=';', encoding='utf-8-sig')

# Fonction pour supprimer les chiffres collés aux mots
def nettoyer_texte(texte):
    if isinstance(texte, str):
        return re.sub(r'\d+', '', texte).strip()
    return texte

# Appliquer la fonction à toutes les colonnes
for col in data.columns:
    data[col] = data[col].apply(nettoyer_texte)

# Remplacer les valeurs manquantes par "Non spécifié"
data.fillna("Non spécifié", inplace=True)

# Afficher un aperçu des données nettoyées
print(data.head())

# Sauvegarder le fichier nettoyé
data.to_csv('fichier_nettoye.csv', index=False, encoding='utf-8-sig')
```

```python
import sqlite3


# Créer une connexion à une base de données SQLite (ici, en mémoire)
conn = sqlite3.connect(':memory:')

# Écrire les données dans une table SQLite
data.to_sql('especes', conn, index=False, if_exists='replace')

# Exemple de requête SQL
query = """
SELECT *
FROM especes
WHERE Taille LIKE '%cm%'
LIMIT 5;
"""

# Exécuter la requête et afficher les résultats
result = pd.read_sql(query, conn)
print(result)

# Fermer la connexion
conn.close()
```
